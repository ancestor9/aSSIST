## **4. 활성화함수**

- 단일층(Single layer), 직선(단순선형회귀식)으로는 아래 그림을 구분할 수없음

- 여러층(Multi layers), 여러 직선(여러 단순선형회귀식)으로는 아래 그림을 구분할 수 있음
- 활성화함수(activation function)의 목적은 인공지능 네트웍에 비선형(Non linearity)을 도입
<img src='%0A'>

**시그모이드 함수(Sigmoid function)**

$${\displaystyle f(x)={\frac {1}{1+e^{-x}}}}
\hspace{1cm}
{\displaystyle f'(x) = f(x)\cdot(1-f(x))}$$

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-6, 6, 100)    
y = sigmoid(x)
y_prime = sigmoid(x)*(1-sigmoid(x))

plt.plot(x, y, label="f (z)")
plt.plot(x, y_prime, label="f '(z)")
plt.plot([0,0],[1.0,0.0], ':', label='th') # 가운데 점선 추가
plt.legend(loc=4, prop={'size': 15}, frameon=True,shadow=True, facecolor="white", edgecolor="black")
plt.axhline(y=0.5, color='blue', linestyle='--')
plt.title('Sigmoid Function')
plt.show()
```

**렐루 함수(ReLU)**

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Relu Function')
plt.show()
```

**소프트맥스 함수(Softamx function)**

소프트맥스 함수는 시그모이드 함수처럼 출력층의 뉴런에서 주로 사용되는데, 세 가지 이상의 (상호 배타적인) 선택지 중 하나를 고르는 다중 클래스 분류(MultiClass Classification) 문제에 주로 사용

```python
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
```

## **7. 딥러닝과 회귀생성모델(regression model)**

```python
x_train= np.random.random((1000,1))
np.mean(x_train), np.std(x_train)
y_train = x_train*2 + np.random.random((1000,1))/3.0

x_test= np.random.random((100,1))
y_test = x_test*2 + np.random.random((100,1))/3.0

plt.plot(x_train, y_train, 'ro')
plt.plot(x_test, y_test, 'bo')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

**7.1. Scikit-learn API의 linear regression**

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error

# 모델구성하기
model = linear_model.LinearRegression()

# 학습하고 예측
y_pred = model.fit(x_train, y_train).predict(x_test)
```

```python
plt.plot(x_test, y_pred, 'go')
plt.plot(x_test, y_test, 'bo')
plt.legend(['prediction', 'test'], loc='upper left')
plt.show()

# 파라미터 추정과 평가
print(f'절편: {model.intercept_}, 기울기:{model.coef_}, mse: {mean_squared_error(y_test, y_pred)}')
```

**7.2. Keras 퍼셉트론 신경망**

```python
from keras.models import Sequential
from keras.layers import Dense
```

입력변수가 1d tensor

```python
# 모델구성
model = Sequential()
model.add(Dense(1, input_dim=1))
```

Parameters (weight, bias)
- The # of weights : {입력행렬 차원(1차원) x layer1 뉴런수(1 nodes)} = 1
- The # of bias : layer1 뉴런 수(1 nodes) = 1
- 추정할 파라미터 개수는 1+1 = 2

```python
model.summary()
```

```python
# 학습하기
model.compile(optimizer='rmsprop', loss='mse')
hist = model.fit(x_train, y_train, epochs=50, batch_size=64) # epoch을 증가시켜보기
```

```python
w, b = model.get_weights()
print(f'절편: {b}, 기울기:{w}')

plt.plot(x_test, y_pred, 'go')
plt.plot(x_test, y_test, 'bo')
plt.plot(x_test, b + w*x_test, 'ro')
plt.legend(['scikit-prediction', 'test', 'perceptron'], loc='upper left')
plt.show()
```

<font color ='red'> epoch을 200으로 증가시키면 예측을 잘한다. 200 epoch을 훈련후 다시 그려보기

```python
# 학습과정 확인하기
hist.history.keys()
plt.plot(hist.history['loss'])
plt.xlabel('epoch'); plt.ylabel('MSE')
```

```python
# 모델 평가
loss = model.evaluate(x_test, y_test, batch_size=30)
print(f'mse: {loss}')
```

```python
# 노드(뉴런) 개수만큼 weight(b, w)가 생성
model.get_weights()
```

```python
plt.plot(x_test, y_pred, 'go')
plt.plot(x_test, y_test, 'bo')
plt.plot(x_test, b + w*x_test, 'ro')
plt.plot(x_test, model.predict(x_test), 'yo')
plt.legend(['scikit-prediction', 'test', 'perceptron', 'fnn'], loc='upper left')
plt.show()
```
